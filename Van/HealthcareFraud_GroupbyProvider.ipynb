```python
def Groupby_Provider(df):
    temp_sum = df.groupby('Provider')[['ClmAdmitDiagnosisCode', 'Inpatient', 'ClmProcedureCode', 'ClmProcedureCode_count',
       'ClmDiagnosisCode_count', 'RenalDiseaseIndicator', 'duplicate', 
       'V5869_diagcode', 'Other_diagcode', 'None_diagcode', 'V5861_diagcode',
       '2724_diagcode', '4011_diagcode', '2449_diagcode', '4019_diagcode',
       '25000_diagcode', '42731_diagcode', 'PHY338032', 'None_PHY',
       'PHY341578', 'PHY357120', 'PHY330576', 'PHY337425', 'PHY314027',
       'PHY327046', 'PHY412132', 'PHY350277', 'PHY423534']].sum().reset_index()
    
    temp_mean = df.groupby('Provider')[['Patient_Connections', 'AttPhys_Connections',
       'InscClaimAmtReimbursed', 'DeductibleAmtPaid','treatment_days','NoOfMonths_PartACov',
       'NoOfMonths_PartBCov', 'IPAnnualReimbursementAmt', 'IPAnnualDeductibleAmt',
       'OPAnnualReimbursementAmt', 'OPAnnualDeductibleAmt', 'num_chronic', 'Gender', 'Age', 'claim_Days', 
       'PotentialFraud']].mean().reset_index()

    temp_count = df.groupby('Provider')['ClaimID'].count().reset_index()

    temp_df = pd.merge(temp_sum, temp_mean, how='outer', on='Provider')

    final_df = pd.merge(temp_count, temp_df, how='outer', on='Provider')

    return final_df
```

```python
final_df = Groupby_Provider(clean_data)
```

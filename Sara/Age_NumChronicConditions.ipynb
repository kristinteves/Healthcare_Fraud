```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
parse_dates = ['ClaimStartDt','DOB','ClaimEndDt','AdmissionDt','DOD','DischargeDt']
data = pd.read_csv('data.csv', parse_dates=parse_dates)
```

## Create Age Column (in years at time of ClaimrStartDt)

```python
data['age'] = data.apply(lambda x: relativedelta(x['ClaimStartDt'], x['DOB']).years, axis=1)
```

#### Drop DOB column

```python
data.drop(['DOB'], inplace=True, axis=1)
```

## Create total number of chronic conditions column

#### Note: This code needs to follow the code that converts 1s and 2s into 1s and 0s in the ChronicCond columns

```python
data['num_chronic'] = data[[x for x in data.columns if "ChronicCond" in x]].sum(axis = "columns")
```

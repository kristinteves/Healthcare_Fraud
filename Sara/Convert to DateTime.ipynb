```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
inpatient = pd.read_csv('Train_Inpatientdata.csv')
outpatient = pd.read_csv('Train_Outpatientdata.csv')
beneficiary = pd.read_csv('Train_Beneficiarydata.csv')
fraud = pd.read_csv('Train_fraud.csv')
```

```python
inpatient['patient_type'] = 1
outpatient['patient_type'] = 2
```

```python
data = inpatient.append(outpatient, sort=False)
```

```python
data = pd.merge(beneficiary, data, how="outer", on="BeneID")
```

```python
data = pd.merge(fraud, data, how="outer", on="Provider")
```

```python
data.head()
```

```python
cols = ['ClaimStartDt','DOB','ClaimEndDt','AdmissionDt','DOD','DischargeDt']
data[cols] = data[cols].apply(pd.to_datetime, format='%Y-%m-%d')
```
